### 1. 뉴스 제목 추출하기

```python
# IT/과학 뉴스 
#url = 'https://news.naver.com/105'

# dict 타입으로 요청 파라미터 설정
req_param = {
    'sid': 105
}
# 
url = 'https://news.naver.com/section/{sid}'.format(**req_param) # ** : 파라이터에 딕셔너리를 전달할 때 사용
print(url)

# 요청 헤더 설정 : 브라우저 정보 (사람처럼 보이게 하기 위함)
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0. Safari/537.36'
}

# requests 의 get() 함수 호출하기 
res = requests.get(url, headers=req_header)
print(res.status_code)
print(res.ok)
print(type(res))
#print(res.text)
# 응답(response)이 OK 이면
# 응답 (response)에서 text 추출
# BeautifulSoup 객체 생성  
if res.ok:
    #pass # if문 내부 구현 예정
    soup = BeautifulSoup(res.text, 'html.parser')
    print(len(soup.select("div.sa_text a[href*='https://n.news.naver.com/mnews/']")))
    # CSS 선택자를 사용해서 a tag 목록 가져오기
    a_tags = soup.select("div.sa_text a[href*='https://n.news.naver.com/mnews/']")
    print(type(a_tags), type(a_tags[0])) #[Tag, Tag]
    
    for a_tag in a_tags:
        title = a_tag.text.strip() # .strip() : 문자열 앞뒤 공백 제거
        link = a_tag['href'] 
        print(title, link)
else:
    # 응답(response)이 Error 이면 status code 출력  
    print(f'Error 발생 : {res.status_code}')
# CSS 선택자

# print(soup.select("div.sa_text a[href*='mnews/article']"))

# <a> 태그 리스트 순회하기    

# 응답(response)이 Error 이면 status code 출력    

```
