```python
import requests
import json
import pandas as pd
from pprint import pprint
```

```python
CLIEENT_ID = 'sdoZ1vLbjVNtY_MSKU07'
CLIENT_SECRET = 'ENe5F_sSpp'
HEADERS = {
    "X-Naver-Client-Id": CLIEENT_ID,
    "X-Naver-Client-Secret": CLIENT_SECRET
}

def search_naver_book(query):

    payload = {
        'query':   query, #'파이썬',
        'display': 100,
        'sort': 'sim'
    }
    url = "https://openapi.naver.com/v1/search/book.json"

    # Naver Book API 요청
    res = requests.get(url, params=payload, headers=HEADERS)

    # 응답 결과 가져오기
    items_data = res.json()['items']
    return items_data
```

```python
# 1. 질문 : 검색어로 찾은 책 목록을 json 파일로 저장하기
query = '파이썬'
items_data = search_naver_book(query)

# pprint로 결과 출력하기
pprint(len(items_data))
items_data = search_naver_book(query)

with open('../data/books.json', 'w', encoding='utf-8') as file:
    json.dump(items_data, file)
```

```python
# 2. 질문 : books.json 파일을 Pandas DataFrame로 저장하기
file_path = '../data/books.json'
df = pd.read_json(file_path)
```

```python
# 3. 질문 : 검색어로 찾은 책 목록 출력하기
df.head()
```

```python
# 4. 질문 : 검색어로 찾은 책 목록 중에서 가격이 2만원 이상인 책만 출력하기
df['discount'] = pd.to_numeric(df['discount'])

high_price_df = df.loc[df['discount'] >= 20000]
high_price_df = high_price_df[['title', 'author', 'discount', 'publisher', 'pubdate']]
high_price_df = high_price_df.sort_values(by='discount', ascending=False).reset_index(drop=True)

high_price_df
```
