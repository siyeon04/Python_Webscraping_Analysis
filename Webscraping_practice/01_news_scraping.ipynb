```python
# reqeusts, bs4 import
import requests
import bs4
# BeautifulSoup 클래스 import
from bs4 import BeautifulSoup

# requests, bs4 버전 확인하기
print(f'requests 버전 = {requests.__version__}')
print(f'bs4 버전 = {bs4.__version__}')
```

```python
# dict 자료형으로 파라미터 설정
res_param = {
    'sid': 'economy'
}

# URL 설정
url = 'https://news.daum.net/{sid}'.format(**res_param)
print(url)

# 헤더 설정
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0. Safari/537.36'
}

# HTTP 요청 및 응답 받기
res = requests.get(url,headers = req_header)

print(type(res)) # 응답 객체 타입 출력
print(res.status_code) # 응답 상태 코드 출력

res.encoding = 'utf-8' # 한글 인코딩 설정

# 응답 본문 출력
if res.ok:
    # BeautifulSoup 객체 생성
    soup = BeautifulSoup(res.text, 'html.parser')

    # CSS 선택자로 뉴스 제목 추출
    li_list = soup.select('ul.list_newsheadline2 li')

    print(f"{type(li_list)}, {len(li_list)}")  # 추출된 뉴스 제목 개수 출력

if li_list:
    #print(type(li_list), type(li_list[0]))  # 추출된 객체 타입 출력

    # 뉴스 제목 & 링크 출력
    for li_tag in soup.select('ul.list_newsheadline2 li'):
        # a 태그 찾기 및 링크 추출
        a_tag = li_tag.find('a')
        link = a_tag['href']
            
        strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
        title = strong_tag.text.strip()

        print(link)
        print(title)
else:
    print(f'Error : {res.status_code}')

print(soup.select("div.cont_thumb"))
```

질문2:  여러개의 section 중 하나를 선택해서 url에서 뉴스기사의 링크와 제목을 출력하는 코드를 함수로 작성하기

경제 뉴스 URL
url = 'https://news.daum.net/economy'

사회 뉴스 URL
url = 'https://news.daum.net/society'

```python
import requests
from bs4 import BeautifulSoup

section_dict = {'기후/환경':'climate','사회':'society','경제':'economy','정치':'politics',\
             '국제':'world','문화':'culture','생활':'life','IT/과학':'tech','인물':'people'}



def print_news(section_name):
    # pass
    sid = section_dict.get(section_name, '경제')
    url = f'https://news.daum.net/{sid}'
    print(f'======> {url} {section_name} 뉴스 <======')

    req_header = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0. Safari/537.36'
    }
    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8' # 한글 인코딩 설정

    if res.ok:
        soup = BeautifulSoup(res.text, 'html.parser')

        for li_tag in soup.select('ul.list_newsheadline2 li')[:1]:
            a_tag = li_tag.find('a')
            link = a_tag['href']

            strong_tag = li_tag.select_one('strong.tit_txt')
            title = strong_tag.text.strip()

            print(link)
            print(title)
    else:
        print(f'Error 발생 : {res.status_code}')
        
print_news('경제')
print_news('사회')
```
