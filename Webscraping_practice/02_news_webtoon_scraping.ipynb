```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import quote_plus
from IPython.display import Image, display
```

```python
def scrape_nate_news():
    # 최신뉴스, 정치, 경제, 사회, 세계, IT/과학 섹션 설정
    sections = {
        '최신뉴스': 'n0100', '정치': 'n0200', '경제': 'n0300', 
        '사회': 'n0400', '세계': 'n0500', 'IT/과학': 'n0600'
    }
    
    base_url = "https://news.nate.com/recent?mid="

    for name, mid in sections.items():
        print(f"\n--- [{name} 섹션] ---")
        url = base_url + mid
        res = requests.get(url)
        soup = BeautifulSoup(res.text, 'html.parser')
        
        # 뉴스 리스트 아이템 선택 (Nate 뉴스 구조에 맞게 선택자 설정)
        articles = soup.select('.mlt01')
        
        for art in articles[:5]:  # 섹션별 5개씩 출력 예시
            title_tag = art.select_one('.lt1 a')
            img_tag = art.select_one('.mType1 img') # 이미지 태그 존재 확인용
            
            if title_tag:
                title = title_tag.get_text().strip()
                link = urljoin(url, title_tag['href'])
                
                print(f"제목: {title}")
                print(f"링크: {link}")
                
                # 이미지 처리
                if img_tag and img_tag.has_attr('src'):
                    img_src = img_tag['src']
                    # //로 시작하는 경로 처리 및 urljoin
                    full_img_url = urljoin(url, img_src)
                    print(f"이미지 URL: {full_img_url}")
                    display(Image(url=full_img_url))
                else:
                    print("이미지 없음")
                print("-" * 30)

# 실행
scrape_nate_news()
```
